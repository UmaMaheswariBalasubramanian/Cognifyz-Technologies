```python
#Analyze the distribution of the target variable ("Aggregate rating") and identify any class imbalances.

target_variable = 'Aggregate rating'

# Analyze the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.histplot(df1[target_variable], kde=True)
plt.xlabel("Target Variable")
plt.ylabel("Frequency")
plt.title("Distribution of Target Variable")
plt.show()
```

```python
# Identify class imbalances for a classification problem
if df1[target_variable].dtype == 'object':
    class_counts = df1[target_variable].value_counts()
    print(class_counts)

    plt.figure(figsize=(8, 6))
    sns.countplot(data=df1, x=target_variable)
    plt.xlabel("Target Variable")
    plt.ylabel("Count")
    plt.title("Class Distribution")
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed
    plt.show()
else:
    print("The target variable is not categorical, it might be a regression problem.")
```

# Descriptive Analysis

Descriptive analysis, also known as descriptive statistics, is a branch of statistics and data analysis that focuses on summarizing and describing the main features of a dataset. Its primary purpose is to provide a clear and concise overview of the data, enabling you to understand its characteristics without making inferences or drawing conclusions about a larger population.


### Problem Statement

Calculate basic statistical measures (mean,
median, standard deviation, etc.) for numerical
columns.

Explore the distribution of categorical
variables like "Country Code","City" and "Cuisines".

Identify the top cuisines and cities with the
highest number of restaurants.

```python
# Calculate basic statistical measures for numerical columns
numeric_columns = df1.select_dtypes(include=[np.number])
statistics = numeric_columns.describe()
```

```python
df1.mean()
```

```python
df1.median()
```

```python
df1.describe()
```

```python
# Explore the distribution of categorical variables
categorical_columns = df1.select_dtypes(include=[np.object])
for col in categorical_columns.columns:
    sns.countplot(data=df1, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed
    plt.show()

# Identify the top cuisines
top_cuisines = df1['Cuisines'].value_counts().head(10)

# Identify the top cities with the highest number of restaurants
top_cities = df1['City'].value_counts().head(10)

print("Basic Statistical Measures for Numerical Columns:")
print(statistics)

print("\nTop 10 Cuisines:")
print(top_cuisines)

print("\nTop 10 Cities with the Highest Number of Restaurants:")
print(top_cities)


```

Geospatial analysis involves working with location-based data, such as latitude and longitude, to gain insights into geographical patterns. To perform geospatial analysis on a dataset of restaurants

### Problrem Statement:

Visualize the locations of restaurants on a
map using latitude and longitude
information.



```python
pip install folium
```

```python
# Create a map centered at a specific location (e.g., city's coordinates or the mean of lat/lon)
restaurant_map = folium.Map(location=[df1['Latitude'].mean(), df1['Longitude'].mean()], zoom_start=12)

# Add markers for each restaurant
for index, row in df1.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(restaurant_map)

# Save or display the map
restaurant_map.save('restaurant_locations_map.html')
```

```python
#To determine if there is any correlation between the restaurant's location (latitude and longitude) 
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Latitude', y='Longitude', hue='Rating color', data=df1)
plt.title("Correlation Between Location and Rating")
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = df1['Latitude'].corr(df1['Rating color'])
print("Correlation Coefficient between Latitude and Rating:", correlation_coefficient)
```
